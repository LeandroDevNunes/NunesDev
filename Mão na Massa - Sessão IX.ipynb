### Valor do documento, valor glosa e valor líquido

O valor do documento representa o valor total da despesa para a qual o parlamentar solicitou a restituição. O valor glosa representa a parcela de retenção de pagamentos para despesas irregulares ou que extrapolam o limite de gastos. O valor líquido representa o montante efetivamente restituído ao parlamentar.

Portanto, iremos considerar somente o valor líquido das despesas, uma vez que ele representa o valor real dos gastos da união com despesas parlamentares.

```python
pt_expenses.loc[pt_expenses['vlrDocumento'] != pt_expenses['vlrLiquido']].head(3)
```

```python
#c) Qual a quantidade de transações (gastos) feitas pelo PT (Partido dos Trabalhadores)?

pt_mean_expenses = pt_expenses['vlrLiquido'].mean()
all_mean_expenses = all_expenses['vlrLiquido'].mean()

print('Quantidade de transações do PT: ', pt_expenses.shape[0])
print('Valor médio das transações do PT: R$ ', round(pt_mean_expenses, 2))
print('Quantidade de transações de todos: ', all_expenses.shape[0])
print('Valor médio das transações de todos: R$ ', round(all_mean_expenses, 2))
```

```python
pt_above_mean_expenses = pt_expenses.loc[pt_expenses['vlrLiquido'] > pt_mean_expenses]
all_above_mean_expenses = all_expenses.loc[all_expenses['vlrLiquido'] > all_mean_expenses]

print('Quantidade de transações do PT acima do valor médio: ', pt_above_mean_expenses.shape[0])
print('Valor médio das transações do PT acima do valor médio: R$ ', round(pt_above_mean_expenses['vlrLiquido'].mean(), 2))
print('Quantidade de transações de todos acima do valor médio: ', all_above_mean_expenses.shape[0])
print('Valor médio das transações de todos acima do valor médio: R$ ', round(all_above_mean_expenses['vlrLiquido'].mean(), 2))
```
