### Função que generaliza a solução dos ítens b, c e d

A função **transform_data** retorna as 15 primeiras linhas de um dataframe de acordo com a sequência de passos abaixo:

1. Para o dataframe **data**, aplique a função lambda **transform**, no eixo da coluna **axis=1** e retorne o resultado em ordem **ascending=True|False**.
2. Armazene o resultado da série resultante do passo 1 (pandas.Series) na variável **result_series**. Esta série contém uma coluna com os valores do resultado da transformação lambda e um índice com a posição da linha em que este valor estava no dataset original antes da ordenação.
3. Localize todas as linhas do dataset original de acordo com o índice **data.loc[result_series.index]**.
4. Armazene na variável **result_data** o resultado do passo 3 somente com as colunas selecionadas em **columns**.
5. Adicione uma nova coluna **result_column** ao resultado; e atribua os valores calculados na transformação **result_series.values**.
6. Do dataset resultante no passo 5 retorne o número de linhas definidos em **rows**.

*Nota: Os valores para a quantidade de linhas no resultado rows=15 e a ordenação descrescente ascending=False foram definidas como padrão e podem ser modificados na chamada da função.*

```python
def transform_data(data, transform, result_column, columns, rows=15, ascending=False):
    result_series = data.apply(transform, axis=1).sort_values(ascending=ascending)
    result_data = data.loc[result_series.index][columns]
    result_data[result_column] = result_series.values
    return result_data.head(rows)
```

**e)** Como é a distribuição de frequência dos municipios dentro de cada faixa de IDHM (<0.6, entre 0.6 e 0.65, 0.66 a 0.75 e >0.75). Como essa distribuição varia de uma região para outra do Brasil?

1. Primeiro criamos uma função que será aplicada como transformação da coluna IDHM para gerar um nova coluna de categorias.

        Faixa de IDHM        Categoria
        abaixo de 0.6	    Ruim
        entre 0.6 e 0.65	 Razoável
        entre 0.66 e 0.75	Bom
        acima de 0.75	    Elevado
        NaN ou nulo	      Desconhecido

```python
def idhm_range(idhm):
    if pd.isnull(idhm):
        return 'Desconhecido'
    elif idhm < 0.6:
        return 'Ruim'
    elif idhm <= 0.65:
        return 'Razoável'
    elif idhm <= 0.75:
        return 'Bom' 
    else:
        return 'Elevado'
```

1 . Em seguida criamos a função que será aplicada como transformação da coluna ESTADO para gerar uma nova coluna REGIAO.

```python
def region(state):
    if state in ['AC', 'AP', 'AM', 'PA', 'RO', 'RR', 'TO']:
        return 'Norte'
    elif state in ['AL', 'BA', 'CE', 'MA', 'PB', 'PE', 'PI', 'RN', 'SE']:
        return 'Nordeste'
    elif state in ['DF', 'GO', 'MT', 'MS']:
        return 'Centro-Oeste'
    elif state in ['ES', 'MG', 'RJ', 'SP']:
        return 'Sudeste'
    elif state in ['PR', 'RS', 'SC']:
        return 'Sul'
    else:
        return 'Desconhecido'
```

1 . Agora criamos uma cópia do dataset original e aplicamos as transformações.

```python
idhm_by_region_data = data[['CIDADE', 'ESTADO', 'IDHM']].copy()
idhm_by_region_data['IDHM_CATEGORIA'] = data['IDHM'].apply(idhm_range)
idhm_by_region_data['REGIAO'] = data['ESTADO'].apply(region)
idhm_by_region_data.head(3)
```

1 . Finalmente calculamos as estatísticas solicitadas no exercício, da seguinte forma:

    . Agrupar os resultados com **groupby** usando as novas colunas criadas **['REGIAO', 'IDHM_CATEGORIA']**
    . Calcular as estatísticas das demais colunas usando **aggregate** com um dicionário que define a função aplicada em cada coluna.
    . Para a **CIDADE** calcule o total **count** de municípios em cada faixa de IDHM.
    . Para o **IDHM** calcule a média **mean** em cada faixa de IDHM.

```python
idhm_by_region_data = idhm_by_region_data.groupby(by=['REGIAO', 'IDHM_CATEGORIA']).aggregate({'CIDADE': 'count', 'IDHM': 'mean'})
idhm_by_region_data
```
